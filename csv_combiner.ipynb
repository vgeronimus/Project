```python
import pandas as pd
import glob

# 1. Find all csv files directly using a pattern
files = glob.glob("participants_data/*.csv")
files

# 2. Create an empty list for the file contents
df_list = []

# 3. Loop through the filenames, read and convert to df
for f in files:
    temp_df = pd.read_csv(f)
    df_list.append(temp_df)
    

# 4. Stack them all together
df = pd.concat(df_list, ignore_index=True)

# 5. See the result
print(df)

# Save file
df.to_csv("combined_data.csv", index=False)
```
